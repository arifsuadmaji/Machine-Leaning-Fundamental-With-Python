```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

print(matplotlib.__version__)
print(np.__version__)
```

# Histogram

```python
mu, sigma = 100, 15 # nilai mean dan nilai standard deviation

# Membangkitkan data berdistribusi normal sebanyak 1000
data =  mu + sigma * np.random.randn(10000)
data
```

```python
len(data)
```

```python
plt.hist(data, bins=50, facecolor='g', alpha=0.50)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, '$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

# Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
# plt.barh(kategori[::-1], data[::-1])

plt.ylabel('Jumlah')
plt.xlabel('Kategori')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
plt.barh(kategori, data)
# plt.barh(kategori[::-1], data[::-1])

plt.grid(linestyle='--', linewidth=1, axis='x', alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

```python
data1 = [25,85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid(linestyle='--', linewidth=1, axis='y',alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

# Pie Plot

```python
plt.pie(data, labels=kategori,startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']

plt.pie(data, labels=kategori,colors=warna,autopct='%1.1f%%',startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

```python
explode_var = [0., 0., 0.2, 0., 0.08]

plt.pie(data, labels=kategori, colors=warna, explode=explode_var, autopct='%1.1f%%', startangle=90)

plt.title('Contoh Exploded Pie Plot')

plt.show()
```

# Boxplot

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Violin Plot

```python
plt.violinplot(data)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Table plot

```python
data_ujian=[['ucup', 80],['bambang', 70],['Tono', 62],['Siti', 77],['Yuni', 87]]
data_ujian
```

```python
table = plt.table(cellText=data_ujian, loc='center')

table.set_fontsize(18)
table.scale(1, 4) # ukuran kolom, baris

ax = plt.gca()
ax.axis(False)

plt.show()
```

# Scatter Plot

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x, y, label='Data 1', color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()

plt.show()
```

## Multiple Scatter Plot

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1, y1, label='Data 1', color='r')
plt.scatter(x2, y2, label='Data 2', color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')

plt.legend()
plt.show()
```
