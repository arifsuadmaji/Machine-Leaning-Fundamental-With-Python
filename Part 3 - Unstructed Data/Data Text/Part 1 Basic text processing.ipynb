```python
import nltk
nltk.download("stopwords")
nltk.download("punkt")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
Text = "Hallo, nama saya Arif. Saya makan bakso harga Rp 15.000 hari ini. Kamu mau?"
```

# Normalization

```python
# Case Folding(merubah text menjadi lower atau upper case)
text = Text.lower()
text
```

# Tokenization

#### Kenapa harus menggunkan tokenizer?<br>
Karena bahasa tidak bisa ketika melihat tanda titik dianggap kalimat, contohnya : rp 15.000, jika dipenggal menggunakan tanda titik 15 dan 000 akan menjadi kalimat yang berbeda.<br>
Jadi, bahasa memiliki rule(aturan)<br>
library punkt telah mempelajari jika ada tanda titik, sebelumnya adalah huruf, dan nextnya adalah spasi maka dipenggal menjadi kalimat, dan aturan bahasa yang lainnya.

```python
# Memenggal/memisahkan kalimat menjadi kalimat yang diinginkan
sent_tokenize(text)
```

```python
# Memenggal/memisahkan per kata
tokens = word_tokenize(text)
tokens
```

# Punctuation Removal/ Alphanumeric Cleansing

```python
punctuation
```

```python
# membuang tanda baca(punctuation)
tokens = [token for token in tokens if token not in punctuation] 
# Untuk setiap token di dalam tokens, jika token tidak ada di dalam punctuation, maka ambil token tersebut
tokens
```

```python
# Alphanumeric Cleansing # Tidak untuk memprediksi harga

tokens = [token for token in tokens if token.isalnum()]
tokens
```

# Stop words removal

```python
stopwords.words("indonesian")
```

dalam kasus nyata kita harus mendefinisikan sendiri(manual) stopwords yang kita inginkan tergantung dengan kondisi

```python
tokens = [token for token in tokens if token not in stopwords.words("indonesian")]
tokens
```

# More Advance Preprocessing :
Oleh sebab itu kita melakukan :
   - Normalization: Menhindari 2 hal yang maknanya sama seperti `saya` dan `Saya`
   - Tokenisasi: menghindari token yang tidak di proses dengan baik misalnya `halo,` atau `ini?`
   - Stemming: menghidari kesamaan makna karena imbuhan seperti `pukul` dan `memukul`
   - Lemmatisasi: mengembalikan kata kebentuk dasarnya seperti `jum'at` menjadi `jumat`
