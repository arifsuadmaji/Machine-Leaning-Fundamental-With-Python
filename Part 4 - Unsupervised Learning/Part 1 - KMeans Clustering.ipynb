# Import Library

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

# Random blobs dataset

```python
# Generate random dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
X
```

# Visualization

```python
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.axis('equal');
```

# K? Berdasarkan data

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X)
centroid = kmeans.cluster_centers_

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=y, s=10)
plt.scatter(centroid[:, 0], centroid[:, 1], c='white', edgecolors='k', s=100, linewidths=2)
plt.axis('equal');
```

# K? Berdasarkan Prediksi

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X)
centroid = kmeans.cluster_centers_

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=label, s=10)
plt.scatter(centroid[:, 0], centroid[:, 1], c='white', edgecolors='k', s=100, linewidths=2)
plt.axis('equal');
```

# K? Elbow Analysis

```python
score = []
for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    label = kmeans.fit_predict(X)
    
    score.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(2, 25), score)
plt.ylabel('Inertia')
plt.xlabel('n_clusters')
```

# K? Silhouette Score

```python
from sklearn.metrics import silhouette_score
```

```python
score = []
for k in range(2, 25):
    kmeans = KMeans(n_clusters=k)
    label = kmeans.fit_predict(X)
    
    siluet = silhouette_score(X, label, metric='euclidean')
    score.append(siluet)
    
plt.figure(figsize=(10,5))
plt.plot(range(2, 25), score, 'r-')
plt.ylabel('Silhouette')
plt.xlabel('n_clusters')
```

# Final KMeans

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X)
centroid = kmeans.cluster_centers_

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=y, s=10)
plt.scatter(centroid[:, 0], centroid[:, 1], c='white', edgecolors='k', s=100, linewidths=2)
plt.axis('equal');
```
